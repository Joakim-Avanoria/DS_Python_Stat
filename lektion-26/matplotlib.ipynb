# Matplotlib

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("resultat_st√§dad.csv")
sumclass = pd.read_csv("summary_by_class.csv")
```

```python
fig, ax = plt.subplots()
ax.plot([0, 1, 2], [1, 2, 1])
ax.set_title("Exempel")
ax.set_xlabel("x")
ax.set_ylabel("y")
ax.grid(True)
plt.tight_layout()
```

```python
fig, ax = plt.subplots()
ax.hist(df["provpo√§ng"].dropna(), bins=10)
ax.set_title("F√∂rdelning av provpo√§ng")
ax.set_xlabel("Po√§ng")
ax.set_ylabel("Antal")
ax.grid(True, axis="y")
plt.tight_layout()
plt.savefig("hist_provpoang.png", dpi=150)
```

```python
fig, ax = plt.subplots()
ax.scatter(df["studietimmar"], df["provpo√§ng"], alpha=0.6)
ax.set_title("Studietimmar vs provpo√§ng")
ax.set_xlabel("Studietimmar")
ax.set_ylabel("Provpo√§ng")
ax.grid(True)
plt.tight_layout()
plt.savefig("scatter_timmar_poang.png", dpi=150)
```

```python
fig, ax = plt.subplots()
ax.bar(sumclass["klass"], sumclass["medel_po√§ng"])
ax.set_title("Medelpo√§ng per klass")
ax.set_xlabel("Klass")
ax.set_xlabel("Medelpo√§ng")
ax.grid(True, axis="y")
plt.tight_layout()
plt.setp(ax.get_xticklabels(), rotation=0)
```
