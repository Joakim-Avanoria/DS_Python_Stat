# Matplotlib

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("resultat_städad.csv")
sumclass = pd.read_csv("summary_by_class.csv")
```

```python
fig, ax = plt.subplots()
ax.plot([0, 1, 2], [1, 2, 1])
ax.set_title("Exempel")
ax.set_xlabel("x")
ax.set_ylabel("y")
ax.grid(True)
plt.tight_layout()
```

```python
fig, ax = plt.subplots()
ax.hist(df["provpoäng"].dropna(), bins=10)
ax.set_title("Fördelning av provpoäng")
ax.set_xlabel("Poäng")
ax.set_ylabel("Antal")
ax.grid(True, axis="y")
plt.tight_layout()
plt.savefig("hist_provpoang.png", dpi=150)
```

```python
fig, ax = plt.subplots()
ax.scatter(df["studietimmar"], df["provpoäng"], alpha=0.6)
ax.set_title("Studietimmar vs provpoäng")
ax.set_xlabel("Studietimmar")
ax.set_ylabel("Provpoäng")
ax.grid(True)
plt.tight_layout()
plt.savefig("scatter_timmar_poang.png", dpi=150)
```

```python
fig, ax = plt.subplots()
ax.bar(sumclass["klass"], sumclass["medel_poäng"])
ax.set_title("Medelpoäng per klass")
ax.set_xlabel("Klass")
ax.set_xlabel("Medelpoäng")
ax.grid(True, axis="y")
plt.tight_layout()
plt.setp(ax.get_xticklabels(), rotation=0)
```
